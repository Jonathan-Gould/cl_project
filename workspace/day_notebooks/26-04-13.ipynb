```python
import cl
from cl import ChannelSet, StimDesign, BurstDesign
from cl.visualisation.jupyter import show_activity
import pathlib
import matplotlib.pyplot as plt
import numpy as np
import time

rng = np.random.default_rng(0)
```

```python
array_side_size = 8
array_size = (array_side_size, array_side_size)
n_channels = np.prod(array_size)
stimable_channels = list(set(range(n_channels)) - {0, 7, 56, 63, 4})
```

```python
array_center = np.array([1, 1]) * (array_side_size - 1) / 2
space_decoder = np.array([np.unravel_index(channel, array_size)[::-1] - array_center for channel in range(array_side_size ** 2)]).T
```

```python
show_activity(mode='2d', use_sidebar=True)
```

```python
n_seconds = 5

xs = [np.array([0,0])]

stim_design = StimDesign(100, -1.5, 100, 1.5)

lead_time_ms = 500

with cl.open() as neurons:

    stim_plan = neurons.create_stim_plan()
    for _ in range(5):
        channels = rng.choice(stimable_channels, size=4, replace=False)
        channels = [int(x) for x in channels]
        stim_plan.stim(ChannelSet(channels), stim_design, lead_time_us=((lead_time_ms*1000) //40) * 40)

    stim_plan.run()
    for tick in neurons.loop(ticks_per_second=60, stop_after_seconds=n_seconds):
        spikes = np.zeros(n_channels)
        for spike in tick.analysis.spikes:
            spikes[spike.channel] += 1
        v = space_decoder @ spikes
        xs.append(xs[-1] + v)

xs = np.array(xs)
```

```python
plt.plot(xs[:,0], xs[:,1])
```
